```julia
df = CSV.read("../data/preprocessed/stock_prices_train.csv",DataFrame,pool=true);
df2 = CSV.read("../data/preprocessed/stock_prices_supplemental.csv",DataFrame,pool=true);
df = vcat(df,df2);
# remove missing 
df = df[completecases(df),:];
println(first(df, 10))
#df = df[df[!,:Date].>=Date.("2017-01-04","yyyy-mm-dd"),:];
df = df[df[!,:Date].<Date.("2022-05-01","yyyy-mm-dd"),:];
# split into train/test
train_split = (df[!, :Date] .< Date.("2022-01-01","yyyy-mm-dd"));
X_train = df[train_split,Not([1,2,7,25])];
X_test = df[.!train_split,Not([1,2,7,25])];
y_train = df[train_split,:Target];
y_test = df[.!train_split,:Target];
```

```julia
using Statistics
n_days = size(unique(df[!, :Date]),1);
n_stocks = size(unique(df[!, :SecuritiesCode]),1);
date_dict = Dict(unique(df[!,:Date]) .=> 1:length(unique(df[!,:Date])));
code_dict = Dict(unique(df[!,:SecuritiesCode]) .=> 1:length(unique(df[!,:SecuritiesCode])));
return_data = zeros(n_days,n_stocks);
for i in 1:size(df,1)
    return_data[date_dict[df[i,:Date]],code_dict[df[i,:SecuritiesCode]]] = df[i,:Target]
end

for i in findall(x->x==0,return_data)
    if i[1]>=2
        return_data[i[1],i[2]] = return_data[i[1]-1,i[2]]
    end
end

# # create return_data with date column
return_data = DataFrame(return_data, :auto);
rename!(return_data, Symbol.(unique(df[!, :SecuritiesCode])));
return_data[!,:Date] = unique(df[!, :Date]);

covariance = cov(return_data[return_data[!,:Date].<Date.("2022-01-04","yyyy-mm-dd"), Not([2001])] |> Tables.matrix);
```

```julia

using Dates, Gurobi, JuMP

function prescription_baseline(df, previous_date, C, D, lambda, sigma)
    Random.seed!(42)
    m = Model(Gurobi.Optimizer)
    set_optimizer_attribute(m, "OutputFlag", 0)

    #extract the part of the dataset that correspond to day 2022/06/23, which is the day before the test day
    idx = df[!,:Date] .== Date(previous_date)
    df_previous_day = df[idx,:]
    
    # extract the 200 performing stocks based on the value of Target for the date "2022-06-23"
    df_previous_day_bottom = sort(df_previous_day, :Target, rev=false)[1:200,:]
    df_previous_day_top = sort(df_previous_day, :Target, rev=true)[1:200,:]
    indeces_top = sortperm(df_previous_day[!,"Target"], rev=true)[1:200]
    indeces_bottom = sortperm(df_previous_day[!,"Target"], rev=false)[1:200]
    println(size(sigma))
    sigma_top = sigma[indeces_top ,: ]
    println(size(sigma_top))
    sigma_top = sigma_top[:,indeces_top]
    println(size(sigma_top))
    sigma_bottom = sigma[indeces_bottom ,:]
    sigma_bottom = sigma_bottom[:,indeces_bottom]
    #we select the top 200 stocks to buy and the bottom 200 stocks to shorts
    @variable(m, w_top[1:200]>=0);
    @variable(m, w_bottom[1:200]<=0);
    @objective(m, Max, sum(w_top[i] * log(1+df_previous_day_top[!,"Target"][i]) for i in 1:200) + sum(w_bottom[i] * log(1+df_previous_day_bottom[!,"Target"][i]) for i in 1:200))
                #+ lambda*(C - (sum(w[i] * df_previous_day_top[!,"Drawdown"][i] for i in 1:200) + sum(w[200+i] * df_previous_day_bottom[!,"Drawdown"][i] for i in 1:200))/600))
    
    #add the constraint that the max of drawdowns on the day 2022/06/23 weighted by w is less that C
    #we are relaxing it as it depends on R
#     @constraint(m, (sum(w[i] * df_2022_06_23_top[!,"Drawdown"][i] for i in 1:200) + sum(w[i] * df_2022_06_23_bottom[!,"Drawdown"][i] for i in 1:200))/600 <= C)
    
    #sum of w should be equal to D
    
    @constraint(m, sum(w_top)+sum(w_bottom) == D)
    @constraint(m, [i in 1:200], w_top[i]<= 0.05)
    @constraint(m, [i in 1:200], -w_bottom[i]<= 0.05)
    @constraint(m, w_top'*sigma_top*w_top<=5)
    @constraint(m, w_bottom'*sigma_bottom*w_bottom<=5)
    optimize!(m)
    #return the value of w
    return value.(w_top), value.(w_bottom)

end
    
```

```julia
w_top_opt_baseline, w_bottom_opt_baseline = prescription_baseline(df, "2022-04-14", 0.01, 1, 1000, covariance)
```

```julia
function oracle_policy(df_day_top, df_day_bottom, C, D, lambda, sigma_top, sigma_bottom)
    Random.seed!(42)
    m = Model(Gurobi.Optimizer)
    set_optimizer_attribute(m, "OutputFlag", 0)


    #we select the top 200 stocks to buy and the bottom 200 stocks to shorts
    @variable(m, w_top[1:200]>=0);
    @variable(m, w_bottom[1:200]<=0);
    @objective(m, Max, sum(w_top[i] * log(1+df_day_top[!,"Target"][i]) for i in 1:200) + sum(w_bottom[200] * log(1+df_day_bottom[!,"Target"][i]) for i in 1:200))
    #+lambda*(C - (sum(w[i] * df_2022_06_24_top[!,"Drawdown"][i] for i in 1:200) + sum(w[200+i] * df_2022_06_24_bottom[!,"Drawdown"][i] for i in 1:200))/600))
    
    #add the constraint that the max of drawdowns on the day 2022/06/23 weighted by w is less that C (we are relaxing it)
    #@constraint(m, (sum(w[i] * df_2022_06_24_top[!,"Drawdown"][i] for i in 1:200) + sum(w[i] * df_2022_06_24_bottom[!,"Drawdown"][i] for i in 1:200))/600 <= C)
    
    #sum of w should be equal to D
    @constraint(m, sum(w_top)+sum(w_bottom) == D)
    @constraint(m, [i in 1:200], w_top[i]<= 0.05)
    @constraint(m, [i in 1:200], -w_bottom[i]<= 0.05)
    @constraint(m, w_top'*sigma_top*w_top<=5)
    @constraint(m, w_bottom'*sigma_bottom*w_bottom<=5)
    optimize!(m)
    #return the value of w
    return value.(w_top), value.(w_bottom)

end
```

```julia
function knn_prescription(y_pred, C, D, lambda,sigma)
    Random.seed!(42)
    m = Model(Gurobi.Optimizer)
    set_optimizer_attribute(m, "OutputFlag", 0) 
    
    #we select the top 200 stocks to buy and the bottom 200 stocks to shorts
    @variable(m, w[1:2000]);
    @objective(m, Max, (1/25)*sum(w[i] * log(1+y_pred[i,j]) for i in 1:2000, j in 1:25))
   # + lambda * (C - (sum(w[i] * drawdown_pred_top[i] for i in 1:200) + sum(w[200+i] * drawdown_pred_bottom[i] for i in 1:200))/600))
    
    #add the constraint that the max of drawdowns on the day 2022/06/23 weighted by w is less that C
    #@constraint(m, (sum(w[i] * drawdown_top[i] for i in 1:200) + sum(w[i] * drawdown_bottom[i] for i in 1:200))/600 <= C)
    
    #sum of w should be equal to D
    @constraint(m, sum(w) == D)
    @constraint(m, [i in 1:2000], -w[i]<= 0.05)
    @constraint(m, [i in 1:2000], w[i]<= 0.05)
    @constraint(m, w'*sigma*w<=5)
    optimize!(m)
    #return the value of w
    return value.(w)

end
```

```julia
function saa_prescription(df_train, C, D, lambda, sigma)
    Random.seed!(42)
    m = Model(Gurobi.Optimizer)
    set_optimizer_attribute(m, "OutputFlag", 0)
    
    y_train = df_train[:, :Target]
    
    #for each stock in df_train, compute the mean of the historical returns over the training set
    y_pred = zeros(2000)
    unique_securities = unique(df_train[:, :SecuritiesCode])
    for i in 1:2000
        y_pred[i] = mean(y_train[df_train[:, :SecuritiesCode] .== unique_securities[i]])
    end
    
    
    @variable(m, w[1:2000]);
    @objective(m, Max, sum(w[i] * log(1+ y_pred[i]) for i in 1:2000))
   # +lambda*(C - (sum(w[i] * drawdown_pred_top[i] for i in 1:200) + sum(w[200+i] * drawdown_pred_bottom[i] for i in 1:200))/600))
    
    #add the constraint that the max of drawdowns on the day 2022/06/23 weighted by w is less that C (we are relaxing it)
    #@constraint(m, (sum(w[i] * df_2022_06_24_top[!,"Drawdown"][i] for i in 1:200) + sum(w[i] * df_2022_06_24_bottom[!,"Drawdown"][i] for i in 1:200))/600 <= C)
    
    @constraint(m, sum(w) == D)
    @constraint(m, [i in 1:2000], -w[i]<= 0.05)
    @constraint(m, [i in 1:2000], w[i]<= 0.05)
    @constraint(m, w'*sigma*w<=5)
    optimize!(m)
    #return the value of w
    return value.(w)

end
```

```julia
w_opt_saa = saa_prescription(df_train, 0.01, 1, 1000, covariance)
```
